# 4. 스타일

```python
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams['font.family'] = 'Malgun Gothic' # Windows
# matplotlib.rcParams['font.family'] = 'AppleGothic'# Mac
matplotlib.rcParams['font.size'] = 15
matplotlib.rcParams['axes.unicode_minus'] = False  # 한글 폰트 사용 시, 마이너스 글자가 깨지는 현상을 해결
```

```python
x = [1, 2, 3]
y = [2, 4, 8]
```

```python
plt.plot(x,y)
```

```python
plt.plot(x,y, linewidth = 5)
```

```python
plt.plot(x, y, marker = 'o')
```

## 마커

```python
plt.plot(x, y, marker = 'o', linestyle = 'None')
```

```python
plt.plot(x, y, marker = 'v')
```

```python
plt.plot(x, y, marker = 'v', markersize = 10)
```

```python
plt.plot(x, y, marker = 'X', markersize = 10)
```

```python
plt.plot(x, y, marker = 'o', markersize = 10, markeredgecolor = 'red')
```

```python
plt.plot(x, y, marker = 'o', markersize = 10, markeredgecolor = 'red', markerfacecolor = "yellow")
```

## 선 스타일

```python
plt.plot(x, y, linestyle = ":")
```

```python
plt.plot(x, y, linestyle = "--")
```

```python
plt.plot(x, y, linestyle = "-.")
```

## 색깔

```python
plt.plot(x, y, color = "pink")
```

```python
plt.plot(x, y, color = "#ff0000")
```

```python
plt.plot(x, y, color = "b")
```

```python
plt.plot(x, y, color = "g")
```

## 포맷

```python
plt.plot(x, y, 'ro--')  # color, markder, linestyle   (red, o, --)
```

```python
plt.plot(x, y, 'bv:')
```

```python
plt.plot(x,y, 'go')  # linestyle이 없으면 선이 안나타남. (linestyle = None)
```

```python
plt.plot(x, y, color = 'g', marker = 'o', linestyle = "None")
```

## 축약어

```python
plt.plot(x, y, marker = 'o', mfc = 'red', ms = 10, mec = "Yellow", ls = ':') 
# mfc = markderfacecolor,  ms = markersize, mec = markderedgecolor,  ls = linestyle
```

## 투명도

```python
plt.plot(x, y, marker = 'o', mfc = 'red', ms = 10, alpha = 0.4)  # alpha = 투명도 (0~1)
```

## 그래프 크기

```python
plt.figure(figsize = (10, 5))
plt.plot(x,y)
```

```python
plt.figure(figsize = (5,10))
plt.plot(x,y)
```

```python
plt.figure(figsize = (10, 5), dpi = 200)  # dots per inch, 확대
plt.plot(x,y)
```

## 배경색

```python
plt.figure(facecolor = "yellow")
plt.plot(x,y)
```

```python
plt.figure(facecolor = "#a1c3ff")
plt.plot(x,y)
```
