# 8. 막대 그래프 (기본)

```python
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams['font.family'] = 'Malgun Gothic' # Windows
# matplotlib.rcParams['font.family'] = 'AppleGothic'# Mac
matplotlib.rcParams['font.size'] = 15
matplotlib.rcParams['axes.unicode_minus'] = False  # 한글 폰트 사용 시, 마이너스 글자가 깨지는 현상을 해결
```

```python
labels = ['강백호', '서태웅', '정대만']
values = [190, 187, 184]

plt.bar(labels, values)
```

```python
labels = ['강백호', '서태웅', '정대만']
values = [190, 187, 184]

plt.bar(labels, values, color = 'r')
```

```python
labels = ['강백호', '서태웅', '정대만']
values = [190, 187, 184]
colors = ['r', 'g', 'b']

plt.bar(labels, values, color = colors, alpha = 0.6)  # alpha : 투명도
```

```python
labels = ['강백호', '서태웅', '정대만']
values = [190, 187, 184]

plt.bar(labels, values)
plt.ylim(175, 195)   # y 축의 데이터를 175 ~ 195로 제한
```

```python
plt.bar(labels, values, width = 0.5)
```

```python
plt.bar(labels, values, width = 0.3)
plt.xticks(rotation = 45)  # x 축의 이름 데이터 각도를 45도로 설정
plt.yticks(rotation = 45)  # y 추그이 키 데이터 각도를 45도로 설정
```

```python
labels = ['강백호', '서태웅', '정대만']
values = [190, 187, 184]
ticks = ['1번학생', '2번학생', '3번학생']

plt.bar(labels, values)
plt.xticks(labels, ticks)
```
